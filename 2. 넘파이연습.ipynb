1. 데이터 분석 프로젝트: 수치형 데이터의 요약, 통계, 시각화
2. 이미지 처리 프로젝트: 이미지 분석
3. 게임, 시뮬레이션 프로그램: 수치 시뮬레이션 등을 구현
4. 머신러닝 / 딥러닝 모델 프로젝트: 금융데이터, 의료데이터, 실험데이터, 센서데이터

```python
import numpy as np

data = [
    [120, 135, 150, 145],  # 제품A
    [100, 110, 140, 130],  # 제품B
    [90, 100, 95, 105] 
]

data
```

```python
profit_per_init = np.array([20, 30, 25])
total_sales = np.dot(profit_per_init, sales_data)
total_sales  # 총 판매금액
```

# 이미지 처리

```python
from PIL import Image
img = Image.open('sample.jpg').convert('RGB')  # 이미지 데이터 읽기 (컬러별 픽셀값)
img_ary = np.array(img)         # array에 넣어줌
new_img = np.clip(img_ary + 50, 100, 200).astype(np.uint8)   # 맘대로 바꿔줌

Image.fromarray(new_img).save('new_sample.jpg')  # array 데이터를 이미지로
```

```python
inverted = 255 - img_ary
Image.fromarray(inverted).save('new_sample1.jpg')
```

# 데이터 시뮬레이션

1차원 배열의 랜덤워크 시뮬레이션

```python
import matplotlib.pyplot as plt

n_step = 1000
step = np.random.choice([-1, 1], size=n_step)  # -1 ~ 1 사이 숫자를 무작위로 1000개 : 행렬로 반환
walk = np.cumsum(step)    # 1000개의 값을 누적

plt.plot(walk)    # 라인차트 그려주는 함수
```

리스트의 슬라이싱과 차이점
- 리스트: 복사본을 리턴
- 배열: 원본 배열의 뷰 -> 복사가 아님, 원본에도 영향을 미침

- 슬라이싱.copy()를 해야 복사가 됨



```python
x = np.arange(10)
x
```

```python
x[:5]
```

```python
new_x = x[4:7]
new_x
```

```python
new_x[0] = 100
new_x, x  # view 뷰의 값을 변경하면 원본의 값도 변경된다!
```

```python
# 원본값을 바꾸지 않으려면 copy() 이용
new_x1 = x[4:7].copy()
new_x1[0] = 200
new_x1, x
```

```python
arr_2d1 = np.arange(16).reshape(4, 4)
arr_2d1, arr_2d1[:2, :2]
```

## 인덱싱 연습하기

```python
# 기준이 될 4x4 배열 생성 (이미지와 동일하게 0부터 15까지의 숫자로 채움)
arr_2d = np.arange(16).reshape((4, 4))

print("원본 4x4 배열:")
print(arr_2d)
print("-" * 30)
```

```python
# 첫 번째 행 전체
arr_2d[0]
```

```python
#마지막 열 전체
arr_2d[:, [-1]]
```

```python
#첫 번째 행과 마지막 행만 선택, 각 행의 첫번째, 마지막 열 선택
arr_2d[::3,::3]
```

```python
# 짝수 인덱스 행(0번째, 2번째 행), 홀수 인덱스 열 선택
arr_2d[::2,1::2]
```

```python
#위의 3차원 행렬을 작성하세요
arr_3d = np.stack([np.arange(0, 36).reshape(6, 6)] * 4)
arr_3d
```

```python
# 2번째 층 선택
arr_3d[1]
```

```python
#네번째행, 두번째 열 원소 선택, 1차원 배열로 작성
arr_3d[:, 3, 1]
```

```python
#배열의 첫 두 층, 마지막 두 행, 마지막 두 열 선택
arr_3d[:2, -2:, -2:]
```

```python
#짝수 인덱스층, 모든행의 세번째열 선택
arr_3d[::2,:,[2]]
```

```python
#세번째 층, 두번째 행에서 네번째 행, 홀수 인덱스 열
arr_3d[[2], 1:4, 1::2]
```
