시각화를 하면 전혀 다른 패턴

```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

그림처럼 만들기

```python
plt.figure(figsize=(7, 4))  # 도화지 (크기조정)
plt.title('5일간 주식 가격')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid()    # 격자 표시
plt.ylim(95, 110)  # y축 범위
plt.xlim(0, len(dates) - 1)

plt.plot(dates, prices, label='Stock Price', marker='o') # plt.plot(x축, y축)
plt.legend()

plt.show()
```

```python
# 여러개 그리기
plt.subplot(1, 2, 1)   # 행은 하나, 컬럼은 두개의 첫번째 그림이다  line
plt.plot(dates, prices, marker='o', color='g')
plt.title('Stock Prices - Subplot 1')
plt.xticks(rotation=40)  # x축 글자 기울여주기 (너무 겹쳐서)

plt.subplot(1, 2, 2)   # 행은 하나, 컬럼은 두개의 두번째 그림이다  bar
plt.bar(dates, prices, color='orange')
plt.title('Stock Prices - Subplot 2')
plt.xticks(rotation=40)

plt.savefig('my_stock_price_chart.png')  # 파일로 저장
```

# 새로운 데이터셋
Seaborn.tips 데이터셋

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
# 1변량 그래프: (히스토그램)
fig = plt.figure(figsize=(8, 5))
axes1 = fig.add_subplot(1, 2, 1)
axes1.hist(data=tips_df, x='total_bill', bins=20)

axes2 = fig.add_subplot(1, 2, 2)
axes2.hist(data=tips_df, x='tip', bins=20)

plt.tight_layout()
plt.show()
```

```python
# boxplot
fig = plt.figure(figsize=(6, 4))
axes = fig.add_subplot(1, 1, 1)
axes.boxplot(data=tips_df, x='total_bill')
plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
axes = fig.add_subplot(1, 1, 1)
axes.boxplot(
    x = [
        tips_df[tips_df['sex'] == 'Female']['total_bill'],
        tips_df[tips_df['sex'] == 'Male']['total_bill']
    ],
    labels=['Female', 'Male']
)

plt.show()
```

```python
tips_df['day'].unique()
```

```python
fig = plt.figure(figsize=(15, 6))
axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

axes1.boxplot(
    x = [
        tips_df[tips_df['day'] == 'Thur']['total_bill'],
        tips_df[tips_df['day'] == 'Fri']['total_bill'],
        tips_df[tips_df['day'] == 'Sat']['total_bill'],
        tips_df[tips_df['day'] == 'Sun']['total_bill']
    ],
    labels=['Thur', 'Fri', 'Sat', 'Sun']
)

axes2.boxplot(
    x = [
        tips_df[tips_df['day'] == 'Thur']['tip'],
        tips_df[tips_df['day'] == 'Fri']['tip'],
        tips_df[tips_df['day'] == 'Sat']['tip'],
        tips_df[tips_df['day'] == 'Sun']['tip']
    ],
    labels=['Thur', 'Fri', 'Sat', 'Sun']
)

axes3.boxplot(
    x = [
        tips_df[tips_df['day'] == 'Thur']['size'],
        tips_df[tips_df['day'] == 'Fri']['size'],
        tips_df[tips_df['day'] == 'Sat']['size'],
        tips_df[tips_df['day'] == 'Sun']['size']
    ],
    labels=['Thur', 'Fri', 'Sat', 'Sun']
)

axes1.set_title('요일별 total_bill')
axes2.set_title('요일별 tip')
axes3.set_title('요일별 size')
plt.show()

# 주말에 청구 금액이 더 큼
# 고액 팁은 토요일과 일요일에 집중되어 있음
```

```python
# scatter plot : 산점도
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.scatter(
    tips_df['total_bill'], tips_df['tip']
)

plt.show()  # 강한 양의 상관관계임을 예측 가능!
```

# 다변량 표현
- 성별에 따라 표현 - 색상
- 값의 크기에 따라 - 점의 크기

```python
colors = {'Female':'orange', 'Male':'blue'}
# scatter plot : 산점도

fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s = tips_df['size'] ** 2 * 10,
    c=tips_df['sex'].map(colors)
)

plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s = tips_df['tip'] ** 2 * 10,
    c=tips_df['sex'].map(colors)
)

plt.show()
```

```python
tips_df['tip_rate'] = tips_df['tip'] / tips_df['total_bill']

fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)

axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s = tips_df['tip_rate'] * 700,
    c = tips_df['sex'].map(colors),
)
plt.show()
```

```python
# 파이 차트 : 비율
# 성별 비율 비교
# 빈도수
sex_counts = tips_df['sex'].value_counts()
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)

axes.pie(
    sex_counts,
    colors=['skyblue', 'pink'],
    labels=sex_counts.index,
    autopct='%.1f%%',
    startangle=90
    )

plt.show()
```

```python
# 다른 변수로 파이차트 그리기
day_counts = tips_df['day'].value_counts()
explode=[0.07, 0.07, 0.07, 0.07]
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)

axes.pie(
    day_counts,
    colors=['lightpink', 'lightyellow', 'lightblue', 'lightgray'],
    labels=day_counts.index,
    autopct='%.1f%%',
    startangle=90,
    explode=explode, 
    shadow=True
    )

plt.show()
```
